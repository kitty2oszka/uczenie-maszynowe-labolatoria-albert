```python
df_copy = df_copy.apply(pd.to_numeric, errors= "ignore").replace({None: pd.NA})
```

```python
missing_idx = df_copy[df_copy.isnull().any(axis=1)].index
print(missing_idx)
```

rodzielam na numeryczne i kategoryczne

```python
df_fill = df_copy.copy()
num_cols = df_copy.select_dtypes(include = "number").columns
kat_cols = df_copy.select_dtypes(exclude = "number").columns
```

uzupełniam wartosci brakujace

```python
for col in num_cols:
  mean_val = df_copy[col].astype(float).mean()
  median_val = df_copy[col].astype(float).median()
  df_fill[f"{col}_mean"] = df_copy[col].fillna(mean_val)
  df_fill[f"{col}_median"] = df_copy[col].fillna(median_val)
```

```python
for col in kat_cols:
    mode_val = df_copy[col].mode(dropna=True)
    if not mode_val.empty:
        df_fill[col] = df_copy[col].fillna(mode_val[0])
```
